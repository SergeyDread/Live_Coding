### Шаг 2.2. Обработка пропусков и приведение к нужным типам

Из таблицы на шаге 1 видно, что для некоторых полей есть пропуски в значениях:
* название и жанр - всего два пропуска, посмотрим на них отдельно

```python
print(data_games.query('name.isna()'))
```

Данные про какие-то неизвестные игры...или игру. Возможно ошибочные, возможно нет. Т.к. записи всего 2 и они не повлияют на результат, не будем учитывать их при дальнейшем анализе.

Похожая ситуация с годом выпуска - заполнять их также не будем, но и не будем учитывать их при формировании общей картины с группировкой по годам. Стоит уточнить, с чем могут быть связаны пропуски: если дело касается также именно старых игр, то вряд ли такая информация поможет в построении прогноза на 2017 (динамичный характер рынка, не смысла смотреть тенденции 20-летней давности).

```python
# удалим строки из набора данных
data_games.dropna(subset=['name', 'year_of_release'], inplace = True)

# сделаем год релиза целыми числами
data_games['year_of_release'] = data_games['year_of_release'].astype(int, errors='ignore')

# проверим изменения
print(data_games.info())
```

Остальные столбцы:
* рейтинг от организации ESRB - скорее всего сама организация появилась не сразу, поэтому и оценки они стали давать играм, вышедшим после определенной даты
* оценка пользователей - количество пропусков почти не отличается от кол-ва кол-ва пустых полей с рейтингом от ESRB. Скорее всего эти данные начали собирать примерно с того же момента
* оценка критиков - судя по примерам значений, пропуски в этом столбце также характерны именно для старых игр

```python
print('Примеры значений для рейтинга от критиков')
print(data_games['critic_score'].value_counts())
print()

print('Примеры значений для рейтинга от ESRB')
print(data_games['rating'].value_counts())
print()

print('Примеры значений для рейтинга от пользователей')
print(data_games['user_score'].value_counts())
print()
```

Для столбца *user_score* характерно большое количество значений tbd, аббревиатуры от To be determined. Возможно оценки игроков на тот момент не были собраны/обработаны, а в рамках текущей задачи это то же самое, что и отсутствие данных.

```python
# приведем тип данных для столбца с оценками игроков к числовому, заменив tbd на NaN
data_games['user_score']=pd.to_numeric(data_games['user_score'], errors='coerce')

# проверка изменений
# print(data_games['user_score'].value_counts())
# print(data_games.info())
```

В случае с рейтингом **0** это тоже оценка, поэтому пропуски логичнее будет заполнить каким-то условным значением, которого в действительности быть не может, но на основе которого в дальнейшем сможем выделить игры "без оценок". Заполним пропуски в столбцах *critic_score*, *user_score* и *rating* значениями "-1" (минус один)

```python
# переменная для значения условного признака
mark = -1

# заполнение пропусков
data_games['critic_score'] = data_games['critic_score'].fillna(mark)
data_games['user_score'] = data_games['user_score'].fillna(mark)
data_games['rating'] = data_games['rating'].fillna(mark)

# проверка изменений
print('Примеры значений для рейтинга от критиков')
print(data_games['critic_score'].value_counts())
print()

print('Примеры значений для рейтинга от ESRB')
print(data_games['rating'].value_counts())
print()

print('Примеры значений для рейтинга от пользователей')
print(data_games['user_score'].value_counts())
print()

# таблица с пропусками по столбцам
print("Кол-во пропусков значений по признакам")
print(data_games.isna().sum())
```

Пропусков больше нет. Прекрасно :)

### Шаг 3.1. Анализ динамики релизов на разных платформах

```python
# построим график всех продаж по годам
data_games_grouped_year = data_games.pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')
data_games_grouped_year = data_games_grouped_year.reset_index()

# print(data_games_grouped_year)

print('График общих продаж по всем платформам')
data_games_grouped_year.plot(x='year_of_release', y='total_sales', figsize=(10,6))
plt.show()
```

Судя по графику, в период где-то после 2000 года рынок игр очень бурно развивался, а к 2012 также стремительно падал по объемам продаж. Взглянем чуть более детально на примере 15 популярных платформ

<font color='blue'> Хорошо, все верно.</font>

```python
# выберем платформы с наибольшими суммарными продажами
top15_platform = data_games.pivot_table(
    index='platform', values='total_sales', aggfunc={'sum','count'}).sort_values(by='sum', ascending=False).head(15)

top15_platform = top15_platform.reset_index()

print(top15_platform.plot(x='platform', y='sum', kind='bar', grid=True, figsize=(12,6)))
# print(top15_platform.plot(x='platform', y='count', kind='bar', grid=True, figsize=(12,6)))
```

<font color='blue'> Платформы определены правильно.</font>


Среди платформ явно выделяются 6 лидеров продаж

```python
# опишем функцию, которая будет строить графики для определенной платформы
def show_lifetime(platform_name):
    df_sales = data_games.query('platform == @platform_name').pivot_table(
        index='year_of_release', values='total_sales', aggfunc=['sum','count'])
    df_sales = df_sales.reset_index()
    
    print('Динамика продаж для платформы', platform_name, 'в деньгах')
    df_sales.plot(x='year_of_release', y='sum', grid=True)
    plt.show()
    print(platform_name, 'в штуках')
    df_sales.plot(x='year_of_release', y='count', grid=True)
    plt.show()
        
    print()
    print()
```

```python
# в цикле посмотрим 15 нужных нам лидеров
for i in range(15):
    show_lifetime(top15_platform.loc[i,'platform'])
```

С визуальной точки зрения, средний **срок жизни платформы** по уровню дохода и по кол-ву выпускаемых на ней игр **около 7 лет**. Есть исключения типа PC, игры на который продавались с момента зарождения индустрии, но для большинства все же характерен именно такой срок.

Следовательно для принятия решения, во что же вкладываться в 2017 году, необходимо выбрать актуальный период, исходя из срока жизни платформ. 2017 минус 7 лет = 2010 год. А значит нам не сильно помогут данные до 2010 и для определения потенциально выгодного направления стоит выбрать платформы, которые после 2010 года показывали положительную динамику по продажам и/или кол-ву выпускаемых игр.

```python
# составим срез данных по платформам для периода с 2010 года
data_games_after2010 = data_games.query('year_of_release >= 2010')
top_platforms_after2010 = data_games_after2010.pivot_table(
    index='platform', values='total_sales', aggfunc={'sum','count'}).sort_values(by='sum', ascending=False)

top_platforms_after2010 = top_platforms_after2010.reset_index()

# а затем также через цикл и функцию графиков посмотрим динамику по платформам
for i in range(len(top_platforms_after2010)):
    show_lifetime(top_platforms_after2010.loc[i,'platform'])
```

На основе графиков можно сделать следующий вывод. Есть несколько потенциально прибыльных платформ:
1. PS4 - старт платформы приходится на 2013 год и исходя из нашей теории про lifetime, должна бы продержаться примерно до 2020 года. Кол-во выпускаемых игр пока только растет, с учетом того, что данные за 2016 неполные. Сумма продаж снижается, однако это общая тенденция всего рынка, а потому не так критична
2. XOne - похожая ситуация. Выручка падает, но кол-во игр на этой платформе пока только растет
3. PC - не подчиняется тренду "срока жизни", видимо, как самая стойкая платформа. И снова снижается выручка, но есть положительная тенденция по кол-ву выпускаемых игр

по PSV - пока спорно, но данные о динамике говорят о том, что скорее всего дальше платформа будет только угасать. Потому не стал относить её к потенциально выгодным

### Шаг 4.1. Определение 5 самых популярных платформ

```python
# функция, выводящия топ 5 значений параметра, у которых самые высокие продажи
def show_plots(parameter):
    
    print('Самые популярные', parameter, 'в Северной Америке')
    df_na = data_games_after2010.pivot_table(
        index=parameter, values='na_sales', aggfunc='sum').sort_values(by='na_sales', ascending=False).head()
    df_na = df_na.reset_index()
    df_na.plot(x=parameter, y='na_sales', kind='bar', grid=True, figsize=(10,5))
    plt.show()
    print()
    
    print('Самые популярные', parameter, 'в Европе')
    df_eu = data_games_after2010.pivot_table(
        index=parameter, values='eu_sales', aggfunc='sum').sort_values(by='eu_sales', ascending=False).head()
    df_eu = df_eu.reset_index()
    df_eu.plot(x=parameter, y='eu_sales', kind='bar', grid=True, figsize=(10,5))
    plt.show()
    print()
    
    print('Самые популярные', parameter, 'в Японии')
    df_jp = data_games_after2010.pivot_table(
        index=parameter, values='jp_sales', aggfunc='sum').sort_values(by='jp_sales', ascending=False).head()
    df_jp = df_jp.reset_index()
    df_jp.plot(x=parameter, y='jp_sales', kind='bar', grid=True, figsize=(10,5))
    plt.show()
    print()
```

```python
show_plots('platform')
```

## Шаг 5. Проверьте гипотезы
Задайте самостоятельно пороговое значение alpha.

Поясните:

### Шаг 5.1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Пороговое значение alpha возьмем стандартное = 5%

Исходя из формулировки самого запроса, легко можно построить нулевую гипотезу: средние двух генеральных совокупностей (в нашем случае данных о рейтинге для разных платформ - поэтому рейтинг будет являться критерием) равны между собой. Следом за ней формулируется альтернативная, обратная: средний рейтинг от пользователей Xbox One значимо отличается от рейтинга пользователей PC

```python
alpha=0.05

# создадим два среза данных, для pc
data_games_pc = data_games.query('platform == "PC" and user_score !=(-1) and year_of_release>=2010')

# и для Xbox One
data_games_xone = data_games.query('platform == "XOne" and user_score !=(-1) and year_of_release>=2010')

results = st.ttest_ind(data_games_pc['user_score'], data_games_xone['user_score'])
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Уровень значимости 98%, что говорит нам о достаточно высокой вероятности того, что оценки от пользователей PC не будут значимо отличаться от оценок игроманов с Xbox One

### Шаг 5.2. Средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта») разные

Также сформулируем нулевую гипотезу: Средние пользовательские рейтинги жанров Action и Sports одинаковые. 
Альтернативная: средний рейтинг игроков для жанра Action значимо отличается от рейтинга жанра Sports. В качестве критерия также берем оценку пользователей *user_score*

```python
# также два среза
data_games_action = data_games.query('genre == "Action" and user_score !=(-1) and year_of_release>=2010')
data_games_sports = data_games.query('genre == "Sports" and user_score !=(-1) and year_of_release>=2010')

results = st.ttest_ind(data_games_action['user_score'], data_games_sports['user_score'])
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Крайне низкий уровень значимости, а следовательно вероятность того, что рейтинги жанров окажутся одинаковыми, крайне мала. 

*(средний пользовательский рейтинг жанра Action действительно отличается от среднего для жанра Sports)*
