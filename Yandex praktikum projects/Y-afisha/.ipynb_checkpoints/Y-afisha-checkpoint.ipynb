```python
# Добавление расчетных столбцов
# даты для дальнейших группировок 
df_visits['day']=df_visits['start_ts'].astype('datetime64[D]')
df_visits['week']=df_visits['start_ts'].astype('datetime64[W]')
df_visits['month']=df_visits['start_ts'].astype('datetime64[M]')

# длительность сессии в минутах
df_visits['duration_ts']=df_visits['end_ts']-df_visits['start_ts']
df_visits['duration_ts']=(df_visits['duration_ts'] / np.timedelta64(1, 'm')).astype('int')
# print(df_visits.head())
```

#### Шаг 2.1.1. Подсчет метрик: Daily active users

```python
# группировка по дням для DAU
dau_total = df_visits.groupby('day').agg({'uid':'nunique'}).reset_index()

# график с изменением метрики
dau_total.plot(x='day', y='uid', kind='line', figsize=(16,5), label='daily active users')
print(dau_total.describe())
```

В целом количество активных пользователей не слишком отличается изо дня в день, однако на графике явно выделяются три выброса:

* в конце наблюдаемого периода

их можно использовать как сигнал для продакт-менеджера и изучить причины выбросов отдельно

Также можно отметить характерную "сезонность" в рамках месяца - пользователи реже заглядывают в сервис в определенные дни недели

```python
# в разбивке по устройствам
dau_device = df_visits.groupby(['day','device']).agg({'uid':'nunique'})
dau_device = dau_device.pivot_table(index='day', columns='device', values='uid', aggfunc='sum')
# print(dau_device)
# график
dau_device.plot(kind='line', figsize=(16,5), label='daily active users')
```

#### Шаг 2.1.6. Подсчет метрик: Retention rate

```python
# вычислим дату первого визита, источник и устройство
df_visits = df_visits.sort_values(by='start_ts')
users = df_visits.groupby('uid').agg({'start_ts':'first', 'source_id':'first', 'device':'first'}).reset_index()
users = users.rename(columns={'start_ts':'first_visit_ts','source_id':'first_source','device':'first_device'})
users['first_visit_date'] = users['first_visit_ts'].astype('datetime64[D]')
users['first_visit_month'] = users['first_visit_ts'].astype('datetime64[M]')

print(users.head())
```

```python
# добавим их в таблицу визитов (отключать для повторных выполнений)
df_visits_new = df_visits.merge(users, on='uid', how='left')

print(df_visits_new.head())
```

```python
# расчет lifetime для события "повторное посещение"
df_visits_new['cohort_visit_lifetime'] = df_visits_new['month'] - df_visits_new['first_visit_month']
df_visits_new['cohort_visit_lifetime'] = df_visits_new['cohort_visit_lifetime'] / np.timedelta64(1, 'M')
df_visits_new['cohort_visit_lifetime'] = df_visits_new['cohort_visit_lifetime'].round().astype('int')
#print(df_visits_new.head())

# Группировка по когортам
cohort_visits = df_visits_new.groupby(['first_visit_month','cohort_visit_lifetime']).agg({'uid':'nunique'}).reset_index()
cohort_visits = cohort_visits.rename(columns={'uid':'cnt_uid'})
#print(cohort_visits.head())

# Создадим датафрейм с описанием величин когорт
initial_visits_count_uid = cohort_visits[cohort_visits['cohort_visit_lifetime']==0][['first_visit_month','cnt_uid']]
initial_visits_count_uid = initial_visits_count_uid.rename(columns={'cnt_uid':'initial_cnt_uid'})
#print(initial_visits_count_uid.head())
```

```python
# и присоединим его к самой таблице с когортами
cohort_visits = cohort_visits.merge(initial_visits_count_uid, on='first_visit_month')

#print(cohort_visits.head())
```

```python
# расчитаем retention rate для каждой
cohort_visits['retention'] = cohort_visits['cnt_uid'] / cohort_visits['initial_cnt_uid']

#print(cohort_visits.head())
```

```python
# выведем сводную таблицу по когортам
cohort_visits_pivot = cohort_visits.pivot_table(index='first_visit_month', columns='cohort_visit_lifetime', values='retention', aggfunc='sum')

sns.set(style='white')
plt.figure(figsize=(16, 8))
plt.title('Visit cohorts: User retention')
sns.heatmap(cohort_visits_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray');
```

```python
print('Усредненный Retenrion rate:',(cohort_visits_pivot.drop(columns=0).mean(axis=0).mean()*100).round(2))
```

#### Шаг 2.2.3. Сколько денег приносят (LTV)

```python
# даты первого заказа уже есть в расчетной таблице users
# посчитаем кол-во новых покупателей для каждого месяца
orders_cohort_sizes = users.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
orders_cohort_sizes = orders_cohort_sizes.rename(columns={'uid':'order_cohort_size'})
#print(orders_cohort_sizes.head())

# сгруппируем сами когорты по дате первого заказа и дате заказа
orders_cohorts = df_orders_new.groupby(['first_order_month','purchase_month']).agg({'revenue':'sum'}).reset_index()
#print(orders_cohorts.head())

# добавим в эту таблицу размер соотв когорт
orders_cohorts = orders_cohorts.merge(orders_cohort_sizes, on='first_order_month')
#print(orders_cohorts.head())
```

```python
# добавим в таблицу возраст когорт
orders_cohorts['age'] = orders_cohorts['purchase_month'] - orders_cohorts['first_order_month']
orders_cohorts['age'] = orders_cohorts['age'] / np.timedelta64(1, 'M')
orders_cohorts['age'] = orders_cohorts['age'].round().astype('int')
#print(orders_cohorts.head(10))

# посчитаем LTV
orders_cohorts['ltv'] = orders_cohorts['revenue'] / orders_cohorts['order_cohort_size']
#print(orders_cohorts.head())
```

```python
# Создадим сводную таблицу для итогов по lifetime value
LTV_pivot = orders_cohorts.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='sum')
LTV_pivot = LTV_pivot.cumsum(axis=1)

plt.figure(figsize=(16, 8))
plt.title('LTV покупателей')
sns.heatmap(LTV_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
```

В выгрузке попались данные за июнь 2018, которые нельзя использовать для выводов. Проигнорируем эти итоги.

По итогам таблицы с lifetime value видно, что наибольший вклад вносят пользователи, которые подключились в сентябре 2017 и июне того же года. А вот LTV клиентов начала 2018 уже так не радует. Возможно из-за сезонного характера бизнеса.

```python
# выберем подходящий "средний" LTV
LTV_mean = LTV_pivot.mean(axis=0)
print(LTV_mean.head(6).round(4))
```

Если возникнет потребность узнать, сколько денег ожидаем получить с привлеченных клиентов, то можно ориентироваться на LTV. Например, на 3-й месяц это 5,82 долл.

#### Шаг 2.3.2. Сколько стоило привлечение одного покупателя из каждого источника

Затраты для каждой пары "день-источник" уже есть в таблице costs. Будем считать, что каждый "первый заказ" был привлечен именно тем источником, с которого соотв клиент в первый раз попал на сайт (таблица visits)

```python
# Создадим когорты для заказов на основе пары день-источник
cohort_size_costs = users.groupby(['first_source','first_order_date']).agg({'uid':'nunique'}).reset_index()
cohort_size_costs = cohort_size_costs.rename(columns={'first_order_date':'dt','first_source':'source_id','uid':'cnt_uid'})
#print(cohort_size_costs.head())
```

```python
# добавляем эту инфу к таблице затрат и считаем расходы на 1 заказ с конкретного источника в конкретный день
df_costs = df_costs.merge(cohort_size_costs, on=['dt','source_id'])
df_costs['acquisition_cost'] = df_costs['costs'] / df_costs['cnt_uid']
#print(df_costs.head())
```

```python
# сделаем срез только нужных столбцов
costs_temp = df_costs[['source_id','dt','acquisition_cost']]
costs_temp = costs_temp.rename(columns={'source_id':'first_source','dt':'first_order_date'})
#print(costs_temp.head())
```

```python
cac_mean_s = costs_temp.groupby('first_source').agg({'acquisition_cost':'mean'})#.reset_index()
print('Средняя стоимость привлечения в разбивке по источникам')
print(cac_mean_s.head(10))
```

Хоть по общим затратам на третий источник и пришлось больше всего, однако за счет количества привлеченных заказов средняя CAC для него в итоге вышла ниже, чем для второго

#### Шаг 2.3.3. Насколько окупились расходы

```python
# добавим в таблицу с данными о платящих пользователях информацию о стоимости их привлечения
users = users.merge(costs_temp, on=['first_order_date','first_source'], how='left')
#print(users.head())
```

```python
# посчитаем кол-во новых покупателей для каждого месяца
orders_cohort_sizes_romi = users.groupby('first_order_month').agg({'uid':'nunique','acquisition_cost':'sum'}).reset_index()
orders_cohort_sizes_romi = orders_cohort_sizes_romi.rename(columns={'uid':'cohort_size'})
orders_cohort_sizes_romi['cac'] = orders_cohort_sizes_romi['acquisition_cost'] / orders_cohort_sizes_romi['cohort_size']
#print(orders_cohort_sizes_romi.head())
```

```python
# добавляем размер когорт в расчеттную таблицу
orders_cohorts_romi = orders_cohorts.merge(orders_cohort_sizes_romi, on='first_order_month')
#print(orders_cohorts_romi.head())
```

```python
# считаем romi
orders_cohorts_romi['ltv'] = orders_cohorts_romi['revenue'] / orders_cohorts_romi['cohort_size']
orders_cohorts_romi['romi'] = orders_cohorts_romi['ltv'] / orders_cohorts_romi['cac']
#print(orders_cohorts_romi.head())
```

```python
# снова создаем срез, чтобы не включать в итоги тех, кто пользуется совсем недавно
orders_cohorts_romi_cut = orders_cohorts_romi.query('first_order_month <= "2018-03-01"')
LTV_pivot_romi = orders_cohorts_romi_cut.pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='sum')
LTV_pivot_romi = LTV_pivot_romi.cumsum(axis=1)

plt.figure(figsize=(16, 8))
plt.title('ROMI покупателей')
sns.heatmap(LTV_pivot_romi, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
```

Пока только две когорты окупили вложения на их привлечение: подключенные в июне 2017 и в сентябре 2017

```python
print(LTV_pivot_romi.mean(axis=0))
```

"Средний" срок окупаемости всех когорт - 8-й месяц

#### ROMI по устройствам

```python
# сделаем срез сразу, чтобы убрать влияние тех, кто пользуется совсем недавно
orders_cohort_sizes_romi_d = users.query('first_order_month <= "2018-03-01"')
#print(orders_cohort_sizes_romi_d.head())
```

```python
# посчитаем кол-во новых покупателей для каждого устройства
orders_cohort_sizes_romi_d = orders_cohort_sizes_romi_d.groupby('first_device').agg({'uid':'nunique','acquisition_cost':'sum'})
orders_cohort_sizes_romi_d = orders_cohort_sizes_romi_d.rename(columns={'uid':'cohort_size'})
orders_cohort_sizes_romi_d['cac'] = orders_cohort_sizes_romi_d['acquisition_cost'] / orders_cohort_sizes_romi_d['cohort_size']
#print(orders_cohort_sizes_romi_d.head())
```

```python
# данные про LTV по устройствам уже есть в таблице LTV_pivot_d, разделим их на cac
orders_cohorts_romi_device = LTV_pivot_d.div(orders_cohort_sizes_romi_d['cac'], axis=0)
#print(orders_cohorts_romi_device) 

plt.figure(figsize=(16, 2))
plt.title('ROMI по устройствам')
sns.heatmap(orders_cohorts_romi_device, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
```

Если смотреть по устройствам, то кажется еще не все вложения себя окупили. Однако у нас есть еще одно измерение "по источникам". Вполне может быть, что какие-то отдельные из них портят общую картинку. Судя по динамике, desktop-устройства вернут наши вложения быстрее.

#### ROMI по источникам

```python
# сделаем срез сразу, чтобы убрать влияние тех, кто пользуется совсем недавно
orders_cohort_sizes_romi_s = users.query('first_order_month <= "2018-03-01"')
#print(orders_cohort_sizes_romi_s.head())
```

```python
# посчитаем кол-во новых покупателей для каждого устройства
orders_cohort_sizes_romi_s = orders_cohort_sizes_romi_s.groupby('first_source').agg({'uid':'nunique','acquisition_cost':'sum'})
orders_cohort_sizes_romi_s = orders_cohort_sizes_romi_s.rename(columns={'uid':'cohort_size'})
orders_cohort_sizes_romi_s['cac'] = orders_cohort_sizes_romi_s['acquisition_cost'] / orders_cohort_sizes_romi_s['cohort_size']
#print(orders_cohort_sizes_romi_s.head())
```

```python
# данные про LTV по устройствам уже есть в таблице LTV_pivot_s, разделим их на cac
orders_cohorts_romi_source = LTV_pivot_s.div(orders_cohort_sizes_romi_s['cac'], axis=0)

# уберем строку про седьмой источник, чтобы она не портила итоги
orders_cohorts_romi_source=orders_cohorts_romi_source.query('first_source != 7')
#print(orders_cohorts_romi_source) 

plt.figure(figsize=(16, 6))
plt.title('ROMI по источникам')
sns.heatmap(orders_cohorts_romi_source, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
```

Действительно, окупились не все источники. Самая печальная статистика по клиентам, привлеченным через источник №3. Не дошли до окупаемости также 4 и 10. А в лидерах определенно первый, девятый и второй

## Шаг 3. Куда маркетологам стоит вкладывать деньги

### Рекомендации по выбору платформы
Рекомендую вкладывать деньги в desktop по нескольким причинам:
- выше LTV - с этих клиентов получаем большую валовую прибыль
- выше ROMI - быстрее возвращаются затраты на инвестиции
- бОльшее количество активных пользователей в день, неделю и месяц (DAU, WAU, MAU)

Прим.: При выборе платформы стоит обращать внимание не только на ROI, но и с каких устройств в первую очередь пользуются продуктом и смотреть на соотношение этих долей, сравнивая с общерыночными тенденциями. Если есть кардинальные отличия, то это может косвенно свидетельствовать о неудобствах/проблемах использования сервиса на конкретной платформе. Если пользоваться априори неудобно, то не стоит ожидать взрывных показателей и для возврата инвестиций.

### С точки зрения выбора источника можно предложить несколько вариантов 
Выбор сделан на основе CAC, LTV и ROMI, т.к. именно они в первую очередь отражают эффективность конкретного канала привлечения.

1. **Source_id 1** - благодаря невысокой стоимости привлечения 1 клиента (CAC) затраты окупились на третий месяц, а сами клиенты приносят значимую часть выручки
2. **Source_id 9** - очень низкая стоимость привлечения, из-за чего, также как и первый источник, быстро окупит вложения, однако LTV у клиентов с этой когорты пока невысокий. Это говорит о том, что мы сможем их дешево привлечь, но в дальнейшем они скорее всего не будут приносить сверхдоходов
3. **Source_id 2** - компромиссный и скорее запасной вариант. Эти клиенты приносят наибольший доход (LTV), но высокая стоимость привлечения сильно подрезает это преимущество. Затраты на привлечение окупаются только на 7-й месяц и далее LTV и ROMI растут крайне медленно
4. Отдельная рекомендация ***не тратить деньги на*** привлечение клиентов через источник ***Source_id 3*** - у него самая высокая стоимость стоимость привлечения и, судя по динамике ROMI, этот источник уже сейчас рискует не окупиться и через три года
